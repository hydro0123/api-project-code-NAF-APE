## Acces to RAW Files

```python
import pandas as pd
import os

files_Programme_BACH=os.path.abspath("Programme-BACH.ipynb")
dossier_notebook = os.path.dirname(files_Programme_BACH)
dossier_project = os.path.dirname(dossier_notebook)
dossier_data=os.path.join(dossier_project,"data")
files_merge1=os.path.join(dossier_data,"Code_APE.xlsx")
files_BACH=os.path.join(dossier_data,"export-bach.csv")

#Import the Code_APE files in data, created by the INSEE_code_APE Notebook
Merge1=pd.read_excel(files_merge1,dtype=str)

#import the BACH data export on the Banque de France website (open source)
Data_BACH = pd.read_csv(files_BACH,sep=";")
```

## Data Cleaning

```python
# Remove the Letter form the section (we are working with Code APE niv2 with only 2 numbers)
Data_BACH["sector"]=Data_BACH["sector"].map(lambda x: x[1:] if len(x)==3 else x)
```

## Files Merging

```python
Data_merge=pd.merge(Merge1,Data_BACH,how="left",left_on="NIV2",right_on="sector")
Data_merge=Data_merge.set_index("NIV5")
```

## Function of the R11

```python
def solvabilite(CodeAPE,TB,FP):
    # TB = Total Bilan 
    # S = Ratio de Solvabilité
    #Solvabilité:
    #
    Q1_solvabilite=Data_merge.loc[CodeAPE,"R11_Q1"]
    Q2_solvabilite=Data_merge.loc[CodeAPE,"R11_Q2"]
    Q3_solvabilite=Data_merge.loc[CodeAPE,"R11_Q3"]
    ratio_de_solvabilite_en_fr=(FP/TB)*100
    ratio_de_solvabilite_en_eu=(TB/FP)*100
    ratio_de_solvabilite_en_fr=round(ratio_de_solvabilite_en_fr,3)
    print("Quartile de ratio EU")
    if ratio_de_solvabilite_en_eu<Q1_solvabilite:
        Q_S_phrase=("Q1: Entreprise fait partie du premier quartile du ratio de solvabilité.")
        Q_S="Q1 solvabilité"
    elif (ratio_de_solvabilite_en_eu>=Q1_solvabilite)&(ratio_de_solvabilite_en_eu<Q2_solvabilite):
        Q_S_phrase=("Q2: Entreprise fait partie du deuxieme quartile du ratio de solvabilité.")
        Q_S="Q2 solvabilité"
    elif (ratio_de_solvabilite_en_eu>=Q2_solvabilite)&(ratio_de_solvabilite_en_eu<Q3_solvabilite):
        Q_S_phrase=("Q3: Entreprise fait partie du troisieme quartile du ratio de solvabilité.")
        Q_S="Q3 solvabilité"
    elif ratio_de_solvabilite_en_eu>=Q3_solvabilite:
        Q_S_phrase=("Q4: Entreprise fait partie du quatrieme quartile du ratio de solvabilité.")
        Q_S="Q4 solvabilité"
    else:
        Q_S_phrase=("Manque de données")
        Q_S="Manque de données"
    Q_S_phrase_FR=("L'entreprise a un ratio de solvabilité (avec Ratio FR) de",ratio_de_solvabilite_en_fr,"%")
    #
    #Label:
    #
    LabelNIV2=Data_merge.loc[CodeAPE,"Label NIV2"]
    LabelNIV1=Data_merge.loc[CodeAPE,"Label NIV1"]
    Label=("L'entreprise fait partie de la section",LabelNIV1,"et l'entreprise est dans la sous-partie",LabelNIV2)
    print("Erreur au niveau du choix du return, insérer 'S' ou 'TB', valeur retourné actuelement par défaut: Ratio de Solvabilité")
    return Q_S + "  |||  " +Q_S_phrase+" (avec Ratio EU)"+ "  |||  " +str(Q_S_phrase_FR)+"  |||  "+str(Label)
solvabilite("71.12B",433509,302005)
```

## Function of the R28

```python
#R28
def R28(CodeAPE,DU,YQ,YR,YS,CF,CG,EE):
    # DU=empruntd et dettes aupres des etablissements de credit
    # YQ= Engagement de credit bail mobilier
    # YR= engagement de credit bail immobilier
    # YS= efets portés a l escompte et non echus
    # CF= disponibilités, brut
    # CG= disponibilités, amortissement, provisions
    # EE= total bilan
    # R28= "Ratio d’endettement net des établissements de crédit"
    Q1_R28=Data_merge.loc[CodeAPE,"R28_Q1"]
    Q2_R28=Data_merge.loc[CodeAPE,"R28_Q2"]
    Q3_R28=Data_merge.loc[CodeAPE,"R28_Q3"]
    ratio_R28_eu=((DU+0.87*YQ+0.65*YR+YS)-(CF-CG))/EE
    ratio_R28_eu=round(ratio_R28_eu,3)
    if ratio_R28_eu<Q1_R28:
        print("Q1: Entreprise fait partie du premier quartile du ratio d'endettement net des établissements de crédit.")
        Q="Q1 endetement"
    elif (ratio_R28_eu>=Q1_R28)&(ratio_R28_eu<Q2_R28):
        print("Q2: Entreprise fait partie du deuxieme quartile du ratio d'endettement net des établissements de crédit.")
        Q="Q2 endetement"
    elif (ratio_R28_eu>=Q2_R28)&(ratio_R28_eu<Q3_R28):
        print("Q3: Entreprise fait partie du troisieme quartile du ratio d'endettement net des établissements de crédit.")
        Q="Q3 endetement"
    elif ratio_R28_eu>=Q3_R28:
        print("Q4: Entreprise fait partie du quatrieme quartile du ratio d'endettement net des établissements de crédit.")
        Q="Q4 endetement"
    else:
        print("Manque de données")
        Q="Manque de données"
    print("L'entreprise a un ratio d'endettement net des établissements de crédit de",str(ratio_R28_eu)," (ratio EU)")

    #
    #Label:
    #
    LabelNIV2=Data_merge.loc[CodeAPE,"Label NIV2"]
    LabelNIV1=Data_merge.loc[CodeAPE,"Label NIV1"]
    print("L'entreprise fait partie de la section",LabelNIV1,"et l'entreprise est dans la sous-partie",LabelNIV2)
    return Q
R28("71.12B",0,0,0,0,99365,0,464153)
```
